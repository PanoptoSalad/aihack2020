```python
def date_to_sum(datestr):
    list1 = datestr.split('/')
    month1 = list1[0]
    if month1[0] == 0:
        month1 == int(month1[1])
    else:
        month1 == int(month1)
        
    date1 = int(list1[1])
    year1 = int(list1[2])
    return(year1,int(month1),date1)
```
