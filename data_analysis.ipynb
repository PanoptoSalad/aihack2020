```python
df_dem = pd.read_csv("dementiadata.csv")[:76]
```

```python
new_total = []
for i in df_dem['total']:
    new_total.append(float(i))
    
new_total = np.array(new_total)
```

```python
percentagedem = df_dem['Sick']/new_total
p_dem = percentagedem
```

```python
year3 = []
for i in df_dem['Year2'][:]:
    if len(i) == 7:
        year3.append((float(i[:2])-1)/12+float(i[-4:]))
    if len(i) == 6:
        year3.append((float(i[:1])-1)/12+float(i[-4:]))
    if len(i) == 4:
        year3.append(float(i))
```

```python
year3
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(year3,p_dem)
plt.ylim(0.003,0.0085)
```
